```python
image = cv2.imread(Input_dir+"\\"+"stop_sign.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

# Laplacian

```python
lap = cv2.Laplacian(image, cv2.CV_64F)
lap = np.uint8(np.absolute(lap))
plt.imshow(lap)
```

# Sobel

```python
sobelX = cv2.Sobel(image, cv2.CV_64F, 1, 0)
sobelX = np.uint8(np.absolute(sobelX))

sobelY = cv2.Sobel(image, cv2.CV_64F, 0, 1)
sobelY = np.uint8(np.absolute(sobelY))

sobelCombined = cv2.bitwise_or(sobelX, sobelY)

fig = plt.figure(figsize=(10, 7))
rows,columns = 2,2

fig.add_subplot(rows, columns, 1)
plt.imshow(image)
plt.axis('off')
plt.title("Original")

fig.add_subplot(rows, columns, 2)
plt.imshow(sobelX)
plt.axis('off')
plt.title("sobelX")
  
fig.add_subplot(rows, columns, 3)
plt.imshow(sobelY)
plt.axis('off')
plt.title("sobelY")
  

fig.add_subplot(rows, columns, 4)
plt.imshow(sobelCombined)
plt.axis('off')
plt.title("sobelCombined")
  
```

# Canny Edge Detection

```python
blur_image = cv2.GaussianBlur(image, (5, 5), 0)
plt.imshow(blur_image)
```

```python
canny = cv2.Canny(blur_image, 30, 150)
canny_image = cv2.cvtColor(canny, cv2.COLOR_BGR2RGB)
plt.imshow(canny_image)
```

```python
blur_image7 = cv2.GaussianBlur(image, (7, 7), 1)
canny7 = cv2.Canny(blur_image7, 20, 50)
canny_image7 = cv2.cvtColor(canny7, cv2.COLOR_BGR2RGB)

blur_image11 = cv2.GaussianBlur(image, (11, 11), 2)
canny11 = cv2.Canny(blur_image11, 15, 200)
canny_image11 = cv2.cvtColor(canny11, cv2.COLOR_BGR2RGB)

fig = plt.figure(figsize=(10, 7))
rows,columns = 2,2

fig.add_subplot(rows, columns, 1)
plt.imshow(blur_image7)
plt.axis('off')
plt.title("7*7 blur filter")

fig.add_subplot(rows, columns, 2)
plt.imshow(blur_image11)
plt.axis('off')
plt.title("11*11 blur filter")

fig.add_subplot(rows, columns, 3)
plt.imshow(canny_image7)
plt.axis('off')
plt.title("canny 7*7 ")

fig.add_subplot(rows, columns, 4)
plt.imshow(canny_image11)
plt.axis('off')
plt.title("canny 11*11 ")
```
